```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import quantstats as qs
from scipy.optimize import minimize

def calculate_average_modified_ma(symbol, period):
    # Fetch historical data from yfinance
    data = yf.download(symbol, start='2006-01-01', end='2023-07-27')

    # Calculate Average Modified MA
    k = 1.0 / period
    data['MMA'] = data['Close'] * k + data['Close'].rolling(window=period).mean() * (1.0 - k)

    return data[['Close', 'MMA']]

def calculate_strategy_returns(data):
    # Determine crossovers
    data['Signal'] = np.where((data['Close'] > data['MMA']) & (data['Close'].shift(1) < data['MMA'].shift(1)), 1, 0)
    data['Signal'] = np.where((data['Close'] < data['MMA']) & (data['Close'].shift(1) > data['MMA'].shift(1)), -1, data['Signal'])

    # Calculate positions and strategy returns
    data['Position'] = data['Signal'].cumsum()
    data['Strategy_Returns'] = data['Close'].pct_change() * data['Position'].shift(1)

    return data

def optimize_parameters(symbol, period_range):
    def objective(period):
        period = int(period)

        data = calculate_average_modified_ma(symbol, period)
        data = calculate_strategy_returns(data)

        strategy_returns = data['Strategy_Returns']
        total_returns = -strategy_returns.sum()

        return total_returns

    result = minimize(objective, period_range.start, bounds=[(min(period_range), max(period_range))])
    best_period = int(result.x)

    return best_period

# Example usage for Nifty 500
nifty500 = '^CRSLDX'     

# Define parameter range to optimize
period_range = range(10, 31)  # Range for MMA period

# Optimize parameter
best_period = optimize_parameters(nifty500, period_range)

# Get the optimized pair of MMA periods
ma1 = best_period
ma2 = int(ma1 * 1.5)#
#multiplier to determine the second MMA period

print("Best Parameters:")
print("MA_1:", ma1)
print("MA_2:", ma2)

# Calculate and plot the optimized strategy
data = calculate_average_modified_ma(nifty500, ma1)
data = calculate_strategy_returns(data)

# ... Rest of the code for generating reports and plotting ...
# Calculate and plot the optimized strategy
data = calculate_average_modified_ma(nifty500, ma1)
data = calculate_strategy_returns(data)

# Create a benchmark (buy and hold Nifty 500 on the first day)
benchmark = pd.DataFrame(index=data.index)
benchmark['Close'] = data['Close']
benchmark['Position'] = 1  # Buy and hold
benchmark['Benchmark_Returns'] = benchmark['Close'].pct_change() * benchmark['Position']

# Backtest the strategy using quantstats
qs.extend_pandas()

# Create a portfolio
portfolio = pd.DataFrame(index=data.index)
portfolio['Close'] = data['Close']
portfolio['Position'] = data['Position']
portfolio['Strategy_Returns'] = data['Strategy_Returns']  # Add the Strategy_Returns column

# Generate the statistics and plots
qs.reports.full(portfolio['Strategy_Returns'], benchmark['Benchmark_Returns'], initial_capital=1000000)
data['Rebalance'] = np.where(data['Signal'] != data['Signal'].shift(1), 1, 0)

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 8), sharex=True)

# Plot Nifty 500 and Average Modified MA
ax1.plot(data.index, data['Close'], color='blue')
ax1.plot(data.index, data['MMA'], color='orange', label='MMA')
ax1.set_ylabel('Nifty 500', color='blue')
ax1.legend()

# Plot Strategy Returns
ax2.plot(portfolio.index, portfolio['Strategy_Returns'], label='Strategy Returns', color='green')
ax2.plot(benchmark.index, benchmark['Benchmark_Returns'], label='Benchmark Returns', color='red')
ax2.set_ylabel('Returns')
ax2.legend()

# Plot Rebalancing Frequency
ax3.plot(data.index, data['Rebalance'], color='purple')
ax3.set_ylabel('Rebalance Frequency')
ax3.set_ylim([0, 1])  # Set y-axis limits to [0, 1] for better visualization

# Plot Number of Times Rebalanced
ax4.plot(data.index, data['Rebalance_Count'], color='orange')
ax4.set_ylabel('Number of Times Rebalanced')

plt.show()
plt.show()

metrics = qs.reports.html(portfolio['Strategy_Returns'], benchmark['Benchmark_Returns'], output='metrics.html', title='Strategy Performance Metrics', initial_capital=1000000)
print("Performance metrics saved as metrics.html")
```
